## Introduction and Problem Statement
The objective of this task is to understand the fundamental steps of data exploration and visualization. We will use the famous **Iris Dataset** to learn how to read, summarize, and visualize data using Python. The goal is to uncover patterns, check data distributions, and identify any outliers in the flower measurements.

## Dataset Understanding and Description
The Iris dataset contains measurements for 150 iris flowers from three different species: Setosa, Versicolor, and Virginica. The features included are:
* Sepal Length (cm)
* Sepal Width (cm)
* Petal Length (cm)
* Petal Width (cm)
* Species (Target Variable)

```python
# Import required libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Loading the Dataset

In this step, we load the Iris dataset using the seaborn library. This dataset is commonly used for learning data analysis and machine learning.

The dataset contains measurements of iris flowers including:

• Sepal Length  
• Sepal Width  
• Petal Length  
• Petal Width  
• Species

```python
# Load iris dataset

df = sns.load_dataset('iris')
```

## Understanding Dataset Structure

Before performing analysis, it is important to examine the dataset structure.

We will check:

• Number of rows and columns using .shape  
• Column names using .columns  
• First few rows using .head()  

This helps us understand how the data is organized.

```python
# Dataset shape

print("Dataset Shape:", df.shape)

# Column names
print("Columns:", df.columns)

# First 5 rows
df.head()
```

## Dataset Features

The dataset contains the following variables:

• sepal_length – Length of the sepal in centimeters  
• sepal_width – Width of the sepal in centimeters  
• petal_length – Length of the petal in centimeters  
• petal_width – Width of the petal in centimeters  
• species – The species of the iris flower

There are three species present in the dataset:

• Setosa  
• Versicolor  
• Virginica

## Scatter Plot Visualization

A scatter plot helps us analyze the relationship between two variables.

Here we compare:

Sepal Length vs Petal Length

This visualization also separates the flower species using different colors, which helps us see how the species differ from each other.

```python
plt.figure(figsize=(6,4))

sns.scatterplot(
    x='sepal_length',
    y='petal_length',
    hue='species',
    data=df
)

plt.title("Sepal Length vs Petal Length")

plt.show()
```

## Histogram Visualization

A histogram shows the distribution of a single variable.

Here we examine how sepal length values are distributed across the dataset. This helps us understand the frequency and spread of the measurements.

```python
plt.figure(figsize=(6,4))

plt.hist(df['sepal_length'], bins=20)

plt.title("Distribution of Sepal Length")

plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()
```

## Box Plot Visualization

A box plot is useful for detecting outliers and understanding the spread of the data.

In this visualization, we compare the petal length distribution across different species of iris flowers.

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    x='species',
    y='petal_length',
    data=df
)

plt.title("Petal Length by Species")

plt.show()
```
